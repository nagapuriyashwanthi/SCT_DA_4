```python
df = pd.read_csv('marketing_campaign.csv', sep=';')
print("Shape:", df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
# Check missing values and data types
print("Missing Values:")
print(df.isnull().sum())
print("\nData Types:")
print(df.dtypes)
print("\nBasic Statistics:")
df.describe()
```

```python
# Check missing values
print("Missing values in Income:", df['Income'].isnull().sum())

# Fix — fill with median income
df['Income'] = df['Income'].fillna(df['Income'].median())

# Confirm fix
print("After fixing:", df['Income'].isnull().sum())
```

```python
# Fix date column
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
print("Date column fixed!")
print(df['Dt_Customer'].dtype)
```

```python
import matplotlib.pyplot as plt

# Chart 1 - Marketing Channel Performance
plt.figure(figsize=(10,6))

channels = ['Web', 'Catalog', 'Store', 'Deals']
purchases = [
    df['NumWebPurchases'].sum(),
    df['NumCatalogPurchases'].sum(),
    df['NumStorePurchases'].sum(),
    df['NumDealsPurchases'].sum()
]

plt.bar(channels, purchases, color=['blue','green','red','orange'])
plt.title('Total Purchases by Marketing Channel')
plt.xlabel('Marketing Channel')
plt.ylabel('Total Purchases')
plt.show()

print("\nChannel Performance:")
for c,p in zip(channels, purchases):
    print(f"{c}: {p}")
```

```python
# Chart 2 - Total Spending by Product Category
plt.figure(figsize=(10,6))

products = ['Wines', 'Fruits', 'Meat', 'Fish', 'Sweets', 'Gold']
spending = [
    df['MntWines'].sum(),
    df['MntFruits'].sum(),
    df['MntMeatProducts'].sum(),
    df['MntFishProducts'].sum(),
    df['MntSweetProducts'].sum(),
    df['MntGoldProds'].sum()
]

plt.bar(products, spending, color=['purple','green','red','blue','pink','gold'])
plt.title('Total Spending by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount Spent')
plt.show()

print("\nSpending by Category:")
for p,s in zip(products, spending):
    print(f"{p}: {s}")
```

```python
# Chart 3 - Campaign Acceptance Rate
plt.figure(figsize=(10,6))

campaigns = ['Campaign 1', 'Campaign 2', 'Campaign 3',
             'Campaign 4', 'Campaign 5', 'Last Campaign']
accepted = [
    df['AcceptedCmp1'].sum(),
    df['AcceptedCmp2'].sum(),
    df['AcceptedCmp3'].sum(),
    df['AcceptedCmp4'].sum(),
    df['AcceptedCmp5'].sum(),
    df['Response'].sum()
]

plt.bar(campaigns, accepted, color='steelblue')
plt.title('Marketing Campaign Acceptance Rate')
plt.xlabel('Campaign')
plt.ylabel('Number of Customers Accepted')
plt.xticks(rotation=45)
plt.show()

print("\nCampaign Acceptance:")
for c,a in zip(campaigns, accepted):
    print(f"{c}: {a} customers")
```

```python
# Chart 4 - Income vs Total Spending
plt.figure(figsize=(10,6))

# Calculate total spending per customer
df['TotalSpending'] = (df['MntWines'] +
                       df['MntFruits'] +
                       df['MntMeatProducts'] +
                       df['MntFishProducts'] +
                       df['MntSweetProducts'] +
                       df['MntGoldProds'])

plt.scatter(df['Income'], df['TotalSpending'],
            alpha=0.5, color='blue')
plt.title('Income vs Total Spending')
plt.xlabel('Customer Income')
plt.ylabel('Total Spending')
plt.show()

print("\nAverage Income:", df['Income'].mean().round(2))
print("Average Spending:", df['TotalSpending'].mean().round(2))
```

```python
# Chart 5 - Education vs Average Spending
plt.figure(figsize=(10,6))

education_spending = df.groupby('Education')['TotalSpending'].mean().round(2)

plt.bar(education_spending.index,
        education_spending.values,
        color=['blue','green','red','purple','orange'])
plt.title('Average Spending by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Spending')
plt.xticks(rotation=45)
plt.show()

print("\nAverage Spending by Education:")
print(education_spending)
```

```python
# Final Business Insights Report
print("="*60)
print("    MARKETING CAMPAIGN - BUSINESS INSIGHTS REPORT")
print("="*60)

print("\n📊 DATASET OVERVIEW:")
print(f"Total Customers: {len(df)}")
print(f"Average Income: ${df['Income'].mean().round(2)}")
print(f"Average Spending: ${df['TotalSpending'].mean().round(2)}")

print("\n📢 MARKETING CHANNEL PERFORMANCE:")
channels = ['Web', 'Catalog', 'Store', 'Deals']
purchases = [
    df['NumWebPurchases'].sum(),
    df['NumCatalogPurchases'].sum(),
    df['NumStorePurchases'].sum(),
    df['NumDealsPurchases'].sum()
]
for c,p in zip(channels, purchases):
    print(f"  {c}: {p} purchases")

print("\n🏆 BEST PERFORMING CHANNEL:")
best = channels[purchases.index(max(purchases))]
print(f"  {best} channel with {max(purchases)} purchases!")

print("\n🛒 PRODUCT SPENDING:")
products = ['Wines', 'Fruits', 'Meat', 'Fish', 'Sweets', 'Gold']
spending = [
    df['MntWines'].sum(),
    df['MntFruits'].sum(),
    df['MntMeatProducts'].sum(),
    df['MntFishProducts'].sum(),
    df['MntSweetProducts'].sum(),
    df['MntGoldProds'].sum()
]
for p,s in zip(products, spending):
    print(f"  {p}: ${s}")

print("\n📣 CAMPAIGN PERFORMANCE:")
campaigns = ['Campaign 1','Campaign 2','Campaign 3',
             'Campaign 4','Campaign 5','Last Campaign']
accepted = [
    df['AcceptedCmp1'].sum(),
    df['AcceptedCmp2'].sum(),
    df['AcceptedCmp3'].sum(),
    df['AcceptedCmp4'].sum(),
    df['AcceptedCmp5'].sum(),
    df['Response'].sum()
]
for c,a in zip(campaigns, accepted):
    print(f"  {c}: {a} customers accepted")

print("\n🎓 TOP SPENDING EDUCATION GROUP:")
top_edu = df.groupby('Education')['TotalSpending'].mean().idxmax()
top_spend = round(df.groupby('Education')['TotalSpending'].mean().max(), 2)
print(f"  {top_edu} with avg spending ${top_spend}")

print("\n💡 RECOMMENDATIONS:")
print("  1. INCREASE budget for Store channel")
print("     → Highest purchases across all channels")
print("  2. INCREASE budget for Web channel")
print("     → Second highest purchases")
print("  3. REDUCE budget for Deals channel")
print("     → Lowest purchases — not effective")
print("  4. TARGET PhD and Master degree customers")
print("     → They spend the most!")
print("  5. FOCUS on Wines and Meat products")
print("     → Highest spending categories")
print("  6. USE Campaign 5 style for future campaigns")
print("     → Highest acceptance rate")

print("\n" + "="*60)
print("         END OF REPORT")
print("="*60)
```
